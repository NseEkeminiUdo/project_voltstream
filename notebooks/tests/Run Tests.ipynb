# Test Suite Runner

Executes pytest with coverage reporting for the `project_volltstream` test suite.

**What this does:**
* Installs pytest and pytest-cov
* Runs all tests in `tests/` directory
* Generates coverage report (terminal + HTML)
* Shows which lines are covered vs missing

```python
import sys
import os

# Prevent bytecode generation
os.environ["PYTHONDONTWRITEBYTECODE"] = "1"
sys.dont_write_bytecode = True

project_dir = "/Workspace/Users/nseekeminiudo@gmail.com/project_volltstream"

print("🧪 Running pytest with coverage...\n")
print("=" * 80)

# Change to project directory
os.chdir(project_dir)

# Run pytest in the same process (can access notebook's Spark session)
import pytest

exit_code = pytest.main(
    ["tests", "--cov=utils", "--cov-report=term-missing", "--cov-report=html", "-v"]
)

print("\n" + "=" * 80)
if exit_code == 0:
    print("✅ ALL TESTS PASSED")
    print(f"📊 HTML Coverage Report: {project_dir}/htmlcov/index.html")
elif exit_code == 5:
    print("⚠️  NO TESTS COLLECTED")
else:
    print(f"❌ TESTS FAILED (exit code: {exit_code})")
print("=" * 80)
```
